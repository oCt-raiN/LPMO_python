```python
# mono nucleotide frequnecy in seq
A=[]
T=[]
G=[]
C=[]
for i in seq_aflotoxin:
  A.append(i.count("A"))
  T.append(i.count("T"))
  G.append(i.count("G"))
  C.append(i.count("C"))
print("The minimum {} and maximum {} of A".format(min(A),max(A)))
print("The minimum {} and maximum {} of T".format(min(T),max(T)))
print("The minimum {} and maximum {} of G".format(min(G),max(G)))
print("The minimum {} and maximum {} of C".format(min(C),max(C)))
```

```python
# di nucleotide frequency in each seq
output = []
for i in seq_aflotoxin:
  di_nuc={}
  for j in range(0,len(i)-1):
    key=i[j]+i[j+1]
    s = str(key)
    #di_nuc[count]={}
    if key in di_nuc.keys():
      di_nuc[s]+=1
    else:
      di_nuc[s]=1
  output.append(di_nuc)
```

```python
import random
from Bio.SeqUtils import GC

con_seq = "ATNCGNNNNNNNACACCNNNTNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNGNGNNGGNNTNNCGNAT"
ran_nuc = ["A","T","C","G"]

output=[]
for i in range(0,5):
  l=""
  for j in con_seq:
    if j=="N":
      sub = random.choice(ran_nuc)
      l=l+sub
    else:
      l=l+j
  if l.count("A")>=min(A) & l.count("A")<=max(A):
    if l.count("T")>=min(T) & l.count("T")<=max(T):
      if l.count("G")>=min(G) & l.count("G")<=max(G):
        if l.count("C")>=min(G) & l.count("C")<=max(C):
          if GC(l)>=min(value) & GC(l)<=max(value):
            if l not in output:
              output.append(l)
  print(i)
```
