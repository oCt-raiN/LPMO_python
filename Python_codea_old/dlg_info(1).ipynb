```python
from operator import index
from ast import Num
import os
import re
import pubchempy as pcp

cid=[]
name=[]
Num_clustering=[]
LBE=[]
LBE_CNFR=[]
LO_CR=[]
LO_Num_CNFR=[]
LO_LBE=[]



fsmile = open("/content/drive/MyDrive/dlgs/smiles.txt")
smile = fsmile.readlines()
#print(smile)
for i in smile:
  l_clust=[]
  lk=i.split()
  f_name = "/content/dlgs_smgrid/"+lk[0]+"+md_0_100_chainA.dlg"
  f_name_dlg = lk[0]+"+md_0_100_chainA.dlg"
  if os.path.isfile(f_name):
    fl = open(f_name)
    k = fl.readlines()
    for m in k:
      q = re.search("^.\s{0,3}\d{1,3}\s\S",m)
      if q :
        l_clust.append(m)
    print(l_clust)
    tmp_lbe=[]
    tmp_n_clust=[]
    tmp_rank=[]
    if len(l_clust)>0:
       for j in range(0,len(l_clust)):
          a =l_clust[j].split()
          tmp_rank.append(int(a[0]))
          tmp_lbe.append(float(a[2]))
          tmp_n_clust.append(int(a[8]))
    cid.append(lk[0])
    try:
       name.append(pcp.Compound.from_cid(lk[0]).synonyms[0])
    except:
       name.append(pcp.Compound.from_cid(lk[0]).iupac_name)
    if len(tmp_rank)>0:
      Num_clustering.append(len(tmp_rank))
      LBE.append(min(tmp_lbe))
      LBE_CNFR.append(tmp_n_clust[tmp_lbe.index(min(tmp_lbe))])
      LO_CR.append(tmp_rank[tmp_n_clust.index(max(tmp_n_clust))])
      LO_LBE.append(tmp_lbe[tmp_n_clust.index(max(tmp_n_clust))])
      LO_Num_CNFR.append(max(tmp_n_clust))
    else :
      Num_clustering.append("Docking Failed")
      LBE.append("Docking Failed")
      LBE_CNFR.append("Docking Failed")
      LO_CR.append("Docking Failed")
      LO_LBE.append("Docking Failed")
      LO_Num_CNFR.append("Docking Failed")
```
