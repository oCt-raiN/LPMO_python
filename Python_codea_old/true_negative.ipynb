```python
from pypdb import *
import Bio
from Bio.PDB import PDBList
import re
id_li =[]
organism =[] 
citation = []
pdbl = PDBList()

for i in range(1,len(unique_l)):
  a = unique_l[i].split()
  all_info = get_info(a[0])
  if all_info:
    id_li.append(a[0])
    citation.append(all_info['citation'][0]["title"])
  else:
    id_li.append(a[0])
    citation.append("cnt find citation")
    pdbl.retrieve_pdb_file(a[0].strip(),pdir='true_negative', file_format="pdb")
  try:
    f = open("/content/true_negative/pdb"+a[0]+".ent")
    fh =f.readlines()
    for j in fh:
      x=re.search("^SOURCE\s+\d\sORGANISM_SCIENTIFIC",j)
      if  x:
        organism.append(j.split(" ",5)[5].strip())
        print(j.split(" ",5)[5].strip())
        break
  except:
    organism.append("cant find organism")
    continue

   
```
