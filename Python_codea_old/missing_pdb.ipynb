```python
#Main -

import json
import pandas as pd
from pypdb import *  
# Opening JSON file
f = open('peptide.json')
  
# returns JSON object as 
# a dictionary
g=[]
data = json.load(f)
b=[]
# Iterating through the json
# list
e=list(data.keys())
for i in e[4234:]:
  s="\""
  #print(data[i][1].split(" (")[0],"\"")
  s+=data[i][1].split(" (")[0]
  s+="\""
  #print(s)
  a=Query(s).search()
  if(a!=None):
    if(len(a)<1000):
      l = []
      l.append(int(i))
      l.append(data[i][1])
      d=[]
      for c in a:
        all_info = get_info(c)
        if(all_info["rcsb_entry_info"]["deposited_polymer_entity_instance_count"]>1):
          g.append(c)
          d.append(c)
      l.append(d)
      b.append(l)
df=pd.DataFrame(b,columns=["PepID","peptide name","pdb id"])
df.to_csv('/content/drive/MyDrive/lab gfv/bamp/PDB.csv', mode='a', index=False, header=False)
t=t+g
#1065-1085
#1169-1181
#1194-1205
#1541
#3276
#3444-3446
#3826-3828
#3870
#4041-4043
#4071-4081
#4147-4155
#4166-4167
#4196-4202
#4207
#4221-4229
#####4234
```
