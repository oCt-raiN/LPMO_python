```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score

# Load the dataset from a CSV file (replace 'your_data.csv' with the name of your CSV file)
data = pd.read_excel('/content/drive/MyDrive/xls/Final_feature_pro.xlsx')

# Separate the features (X) and the target variable (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Preprocess the data by standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the range of learning rate and regularization (alpha) values
learning_rates = np.arange(0.001, 0.02, 0.005)
alphas = np.arange(0.001, 0.02, 0.005)

# Initialize an empty F1 score matrix
f1_matrix = np.zeros((len(learning_rates), len(alphas)))

# Train neural networks with different learning rates and regularization (alpha) values
for i, lr in enumerate(learning_rates):
    for j, alpha in enumerate(alphas):
        nn_classifier = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam',
                                      learning_rate_init=lr, alpha=alpha, random_state=42, max_iter=1000)
        nn_classifier.fit(X_train_scaled, y_train)
        y_pred = nn_classifier.predict(X_test_scaled)
        f1 = f1_score(y_test, y_pred)
        f1_matrix[i, j] = f1


# Create a colorblind-friendly multi-color palette
multicolor_palette = sns.color_palette("husl", 6)
multicolor_cmap = sns.blend_palette(multicolor_palette, as_cmap=True)


# Create a heatmap using the F1 score matrix
plt.figure(figsize=(10, 8))
sns.heatmap(f1_matrix, annot=False, cmap=multicolor_cmap, xticklabels=alphas, yticklabels=learning_rates,
            cbar_kws={'label': 'F1 Score'})
plt.xlabel('Alpha')
plt.ylabel('Learning Rate')
plt.title('F1 Score Matrix for Neural Networks with Different Hyperparameters')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, confusion_matrix, precision_recall_curve

# Load the dataset from a CSV file (replace 'your_data.csv' with the name of your CSV file)
data = pd.read_excel('/content/drive/MyDrive/xls/Final_feature.xlsx')

# Separate the features (X) and the target variable (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Preprocess the data by standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create the Multi-Layer Perceptron (Neural Network) classifier
nn_classifier = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', random_state=42,alpha=0.001,learning_rate_init=0.001)

# Train the classifier
nn_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred_proba = nn_classifier.predict_proba(X_test_scaled)[:, 1]

# Calculate the ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Calculate the Precision-Recall curve and AUC
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
pr_auc = auc(recall, precision)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, label='Neural Network (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Plot the Precision-Recall curve
plt.figure()
plt.plot(recall, precision, label='Neural Network (AUC = %0.2f)' % pr_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower right")
plt.show()

y_pred = (y_pred_proba > 0.5).astype(int)
# Compute and print the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)


plt.imshow(cm, cmap='Blues')
plt.title('Confusion matrix')
plt.colorbar()
plt.xticks([0,1], ['Negative', 'Positive'])
plt.yticks([0,1], ['Negative', 'Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
