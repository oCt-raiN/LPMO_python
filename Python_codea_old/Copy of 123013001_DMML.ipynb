```python
# A.Aathi Manikandan
# 123013001

#importing math library
import math
n=int(input("Enter the order for Minkowsi Distance\n"))
a=[(1,2),(2,3),(7,6),(3,4)]
print()

#initialization of matrix
m = [[0 for c in range(len(a))] for d in range(len(a))] 

#Calculation of morkowsi distance
for i in range(len(a)):
  for j in range(len(a)):
    s=abs((a[i][0]-a[j][0])**n)+abs((a[i][1]-a[j][1])**n)
    if(n==1):
      m[i][j]=s
    elif(n==2):
      m[i][j]=math.sqrt(s)
    else:
      m[i][j]=s**(1/n)

#Uppertriangular matrix
for i in range(len(a)):
  for j in range(len(a)):
    if(i>j):
      m[i][j]=0

#Displaying the result
for r in m:
  print(r)
```
