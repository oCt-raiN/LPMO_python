```python
import re
import pandas as pd
df = open('AA10.txt', "r")
lines = df.readlines()
df.close()
print(lines)
```

```python
b=[]
for i in range(0,len(lines)):
  a=lines[i].split("\t")
  b.append(a[2])
```

```python
unique_list = []
# traverse for all elements
for x in b:
# check if exists in unique_list or not
  if x not in unique_list:
    if(x.split(" ")[1]!="sp."):
      unique_list.append(x.split(" ")[0]+" "+x.split(" ")[1])
```

```python
unique_list1 = []
# traverse for all elements
for x in unique_list:
# check if exists in unique_list or not
  if x not in unique_list1:
    unique_list1.append(x)
```

```python
len(unique_list1)
```

```python
import pandas as pd
import re

q = []

for i in unique_list1:
  e = list(data["data"].keys())
  for j in e[1:]:
    c = str(j)
    h = re.compile("|".join(data["data"][c].keys()))
    if h.search("organisms"):
      for k in data["data"][c]['organisms']:
        g=[]
        if(i==data["data"][c]["organisms"][k]["value"]):
          g.append(i)
          g.append(data["data"][j]["name"])
          g.append(j)
        if(len(g)==3):
          q.append(g)
          df=pd.DataFrame(q,columns=["AA10","enz_name","EC_num"])
  print(i)
```
