```python
import re
# f8 = open("log.txt","a")

for i in range(0,len(AA9)):#len(AA9)
  f_name = "/content/AA9/pdb"+AA9[i].strip().lower()+".ent"
  faf = open("/content/drive/MyDrive/ML/existing_struc/structures/AA9/"+"log_AA9.txt","a")
  try:
    fh = open(f_name,"r")
    # f8.write(f_name)
    # f8.write("\n")
    fl_li = fh.readlines()
    O_name = "/content/drive/MyDrive/ML/existing_struc/structures/AA9/"+AA9[i].strip().lower()+".pdb"
    fh1 = open(O_name,"a")
    for j in range(0,len(fl_li)):
      x =re.search("^ATOM",fl_li[j])
      if x:
        a = fl_li[j].split()
        if a[4] == "A" :
          fh1.write(fl_li[j])
          # fh1.write("\n")
        # f8.write(str(len(a)))
        # f8.write("\n")
  except:
    faf.write(AA9[i])
    continue
```
