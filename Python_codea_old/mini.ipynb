```python
import re
import pandas as pd
df = open('AA10.txt', "r")
lines = df.readlines()
df.close()
print(lines)
```

```python
b=[]
for i in range(0,len(lines)):
  a=lines[i].split("\t")
  b.append(a[2])
```

```python
unique_list = []
# traverse for all elements
for x in b:
# check if exists in unique_list or not
  if x not in unique_list:
    if(x.split(" ")[1]!="sp."):
      unique_list.append(x.split(" ")[0]+" "+x.split(" ")[1])
```

```python
unique_list1 = []
# traverse for all elements
for x in unique_list:
# check if exists in unique_list or not
  if x not in unique_list1:
    unique_list1.append(x)
```

```python
len(unique_list1)
```

```python
with open("AA10_fi.txt",'w') as f:
  for x in unique_list1:
    f.write(x + "\n")
```
