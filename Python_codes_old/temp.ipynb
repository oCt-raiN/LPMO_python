```python
import json
import pandas as pd
from pypdb import *  
# Opening JSON file
f = open('peptide.json')
  
# returns JSON object as 
# a dictionary
g=[]
data = json.load(f)
b=[]
# Iterating through the json
# list
e=list(data.keys())
for i in e[0:len(e)-1]:
  s="\""
  #print(data[i][1].split(" (")[0],"\"")
  s+=data[i][1].split(" (")[0]
  s+="\""
  #print(s)
  a=Query(s).search()
  if(a!=None):
    if(len(a)<1000):
      l = []
      l.append(int(i))
      l.append(data[i][1])
      d=[]
      for c in a:
        all_info = get_info(c)
        if(all_info["rcsb_entry_info"]
           ["deposited_polymer_entity_instance_count"]
           ==2):
          g.append(c)
          d.append(c)
      l.append(d)
      b.append(l)
df=pd.DataFrame(b,columns=["PepID","peptide name","pdb id"])
df.to_csv('/content/drive/MyDrive/lab gfv/bamp/pdb_chains_2.csv', mode='a', index=False,
          header=False)
t=t+g
```
