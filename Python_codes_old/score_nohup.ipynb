```python
import pandas as pd
import re

cid =[]
pLDDT =[]
ptmscore =[]

fh = open("AA11_rank1.txt")
f1 = fh.readlines()
id = []
fh1 = open("nohup.out",encoding='utf-8')
f2 =fh1.readlines()
model_no =[]
for i in f1:
    a =i.split("_",4)
    id.append(a[0])
    model_no.append(a[-1].strip())
# print(model_no)


for i in range(0,len(f2)):#len(f2)
    x = re.search("\d+\-\d+\-\d+\s\d+\:\d+\:\d+\,\d+\sQuery",f2[i])
    if x :
        cid_get = f2[i].split()[-3]
        temp_index = id.index(cid_get)
        reg = "\d+\-\d+\-\d+\s\d+\:\d+\:\d+\,\d+\s"+model_no[temp_index]
        for j in range(i,len(f2)):
            y = re.search(reg,f2[j])
            if y:
                cid.append(cid_get)
                pLDDT_temp = float(f2[j].split()[-4])
                ptmscore_temp = float(f2[j].split()[-1])
                pLDDT.append(pLDDT_temp)
                ptmscore.append(ptmscore_temp)
                break

df = pd.DataFrame(list(zip(cid,pLDDT,ptmscore)),columns=["cid","pLDDT","ptmscore"])
df.to_csv("A11_score.csv")
```
