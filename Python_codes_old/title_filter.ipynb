```python
import re

q=[]
for i in range(0,len(a)):
  x=re.search(a["peptide"][i].split(" ")[0].lower(),a["title"][i])
  c=[]
  if(x!=None):
    c.append(a["pep_id"][i])
    c.append(a["peptide"][i])
    c.append(a["pdb_id"][i])
    c.append(a["chains"][i])
    c.append(a["title"][i])
    q.append(c)
    df=pd.DataFrame(q,columns=["PepID","peptide","pdb_id","chains","title"])
```
