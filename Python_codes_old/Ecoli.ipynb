```python
import numpy as np
import pandas as pd 


#load the E coli Corte test model 
import cobra
model = cobra.test.create_test_model("textbook")

def o2_dependent_flux(bound):
  """Given an o2 flux in bound , return the corresponding pFBA
  solution
  """
  #use a cabrapy context mandager to hanle the reversible manipulation of model 
  #bound
  with model:
    model.reactions.EX_o2_e.lower_bound = -bound
    return cobra.flux_analysis.pfba(model).fluxes

def o2_dependent_fva(bound):
  """Given an o2 flux in bound , return the corresponding Fva bound
  """
  with model:
    model.reactions.EX_o2_e.lower_bound = -bound
    fva = cobra.flux_analysis.flux_variability_analysis(
        model, reaction_list=model.exchanges,
        fraction_ofoptimum=0.99).reset_index()
    fva['o2_bound'] = bound
    return fva

#define a list of oxygen flux bounds to iterate over 
o2_bound =np.limspace(0,30,200)

#save the lsit of FVA solutions to pandas DAtaframe
o2_fluxes = pd.concat([o2_dependent_flux(bound) for bound in o2_bound],
                      axis=1,ignore_index=True).T
o2_fva = pd.concat([o2_dependent_fva(bound) for bound in o2_bound],
                   axis=0)
```
