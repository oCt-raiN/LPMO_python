```python
import pandas as pd

df = pd.read_csv("full.csv")
```

```python
df
```

```python
from tqdm import tqdm
import re

Pep_id=[]
Dramp_id=[]
Name =[]

Pep_id_N=[]
Dramp_id_N=[]
Name_N=[]
for i, row in tqdm(df.iterrows()):
  x =None
  y= None
  txt =  str(df["Activity"][i])
  x = re.search("Anti-Gram+", txt)
  y = re.search("Anti-Gram-",txt)
  if x and y:
    Pep_id.append(df["PepID"][i])
    Pep_id_N.append(df["PepID"][i])
    Dramp_id.append(df["DRAMP_ID"][i])
    Dramp_id_N.append(df["DRAMP_ID"][i])
    Name.append(df["Name"][i])
    Name_N.append(df["Name"][i])
  elif x:
    Pep_id.append(df["PepID"][i])
    Dramp_id.append(df["DRAMP_ID"][i])
    Name.append(df["Name"][i])
  elif y:
    Pep_id_N.append(df["PepID"][i])
    Dramp_id_N.append(df["DRAMP_ID"][i])
    Name_N.append(df["Name"][i])

df1 = pd.DataFrame(list(zip(Pep_id,Dramp_id,Name)),columns=["PepID","DRAMP_ID","Name"])
df2 = pd.DataFrame(list(zip(Pep_id_N,Dramp_id_N,Name_N)),columns=["PepID","DRAMP_ID","Name"])
```
