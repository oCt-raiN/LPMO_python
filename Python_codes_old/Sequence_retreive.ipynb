```python
import re

def sequence(q):
  x=None
  n=q
  f=open(n,"r")
  for lines in f:
    x= re.search("^[A-Z]",lines)
    if(x!=None):
      return lines
```

```python
import re

def sequence1(q):
  x=None
  n=q
  sr=[]
  f=open(n,"r")
  for lines in f:
    x= re.search("^[A-Z]",lines)
    if(x!=None):
      r = lines
      sr.append(r)
  return sr
```

```python
import requests
 
def fasta(n):
  s="https://raw.githubusercontent.com/KarishmaKaushikLab/B-AMP/master/static/peptides/fasta/"
  s+=n
  ##print(s)
  q=n
  ##print(q)
  respond = requests.get(s)
  ##print(respond.content)
  open(q,"wb").write(respond.content)
```
