```python
l =  list(bamp["DRAMP_ID"])
```

```python
del l[0:2]
```

```python
keys_e = ['DRAMP_ID', 'Sequence', 'Sequence_Length', 'Name', 'Swiss_Prot_Entry',
       'Family', 'Gene', 'Source', 'Activity', 'Protein_existence',
       'Structure', 'Structure_Description', 'PDB_ID', 'Comments',
       'Target_Organism', 'Hemolytic_activity', 'Linear/Cyclic/Branched',
       'N-terminal_Modification', 'C-terminal_Modification',
       'Other_Modifications', 'Stereochemistry', 'Cytotoxicity',
       'Binding_Traget', 'Pubmed_ID', 'Reference', 'Author', 'Title']
```

```python
output_af=[]
for i in range(0,len(anti_f["DRAMP_ID"])):
  k = []
  key = anti_f["DRAMP_ID"][i]
  if key not in l:
    for j in keys_e:
      k.append(anti_f[j][i])
    output_af.append(k)
```

```python
len(output_af)
```

```python
output_ab=[]
for i in range(0,len(anti_b["DRAMP_ID"])):
  k = []
  key = anti_b["DRAMP_ID"][i]
  if key not in l:
    for j in keys_e:
      k.append(anti_b[j][i])
    output_ab.append(k)
```

```python
len(output_ab)
```

```python
output_ag=[]
for i in range(0,len(anti_g["DRAMP_ID"])):
  k = []
  key = anti_g["DRAMP_ID"][i]
  if key not in l:
    for j in keys_e:
      k.append(anti_g[j][i])
    output_ag.append(k)
```

```python
len(output_ag)
```

```python
df = pd.DataFrame(output_ab,columns=keys_e)
df1 = pd.DataFrame(output_af,columns=keys_e)
df2 = pd.DataFrame(output_ag,columns=keys_e)
```

```python
h = list(df["DRAMP_ID"])
for i in range(0,len(df1["DRAMP_ID"])):
  k = []
  key = df1["DRAMP_ID"][i]
  if key not in h:
    for j in keys_e:
      k.append(df1[j][i])
    output_ab.append(k)
```

```python
for i in range(0,len(df2["DRAMP_ID"])):
  k = []
  key = df2["DRAMP_ID"][i]
  if key not in h:
    for j in keys_e:
      k.append(df2[j][i])
    output_ab.append(k)
```

```python
len(output_ab)
```
