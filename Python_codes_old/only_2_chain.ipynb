```python
import pandas as pd
from pypdb import *  
#for i in range(0):
b=[]
q=[]
p=[]
output=[]
for n in range(0,len(a)):
  b=a['pdb_id'][n]
  c=[]
  c.append(a['pep_id'][n])
  c.append(a['pepdide'][n])
  all_info = get_info(b)
  if(all_info["rcsb_entry_info"]["deposited_polymer_entity_instance_count"]!=0):
    c.append(a['pdb_id'][n])
    p=all_info["rcsb_entry_info"]["deposited_polymer_entity_instance_count"]
    c.append(p)
    output.append(a['pdb_id'][n])
  else:
    c.append('[0]')
  q.append(c)
  df=pd.DataFrame(q,columns=["PepID","peptide","pdb_id","No_chains"])
  print(df)
```
